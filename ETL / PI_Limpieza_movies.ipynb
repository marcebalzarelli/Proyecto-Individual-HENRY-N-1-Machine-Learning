```python
# Convierto la columna release_date a tipo de dato datetime
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Filtro las filas con fechas válidas
df_valid_dates = df[df['release_date'].notnull()]
```

```python
# Creo la columna release_year extrayendo el año de la fecha de estreno
df_valid_dates['release_year'] = df_valid_dates['release_date'].dt.year
```

```python
df_valid_dates.head(3)
```

```python
# Combino los resultados en un nuevo DataFrame
df_final = pd.concat([df[df['release_date'].isnull()], df_valid_dates])
```

```python
print(df_final)
```

```python
# Convierto la columna "budget" al tipo numérico
df_final['budget'] = pd.to_numeric(df_final['budget'], errors='coerce')

# Creo la columna "return" en el DataFrame df_final
df_final['return'] = df_final.apply(lambda row: row['revenue'] / row['budget'] if row['revenue'] != 0 and row['budget'] != 0 else 0, axis=1)

```

```python
df_final.head(3)
```

```python
# Elimino las filas con valores nulos en una columna release_year
column_name = 'release_year'
df_final = df_final.dropna(subset=[column_name])
```

```python
# Cambio el formato de la columna release_year a entero
df_final['release_year'] = df_final['release_year'].astype(int)
```

```python
df_final.head(3)
```

```python
# Algunos campos estan anidados y deben ser desanidados para unirlos al dataset
# Para las listas

def fetch_name(obj):
    if isinstance(obj,str) and '{' in obj:
        s = []
        for i in ast.literal_eval(obj):
            s.append(i['name']);
        return s
```

```python
# Para los diccionarios

def fetch_name1(obj):
    if isinstance(obj,str) and '{' in obj:
        dic = ast.literal_eval(obj)
        return dic['name']
```

```python
df_final['genres'] = df_final['genres'].apply(fetch_name)
df_final['belongs_to_collection'] = df_final['belongs_to_collection'].apply(fetch_name1)
df_final['production_companies'] = df_final['production_companies'].apply(fetch_name)
df_final['production_countries'] = df_final['production_countries'].apply(fetch_name)
df_final['spoken_languages'] = df_final['spoken_languages'].apply(fetch_name)
```

```python
# Filtro las filas con fechas válidas
df_valid_dates = df_final[df_final['release_date'].notnull()]
```

```python
# Agrego una nueva columna 'DiaSemana' que contiene el día de la semana correspondiente a cada fecha
df_final['DiaSemana'] = df_final['release_date'].dt.day_name()
```

```python
# Reemplazo los nombres de los días en inglés por los correspondientes en español
nombres_dias_ingles = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
nombres_dias_espanol = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
df_final['DiaSemana'] = df_final['DiaSemana'].replace(nombres_dias_ingles, nombres_dias_espanol)
```

```python
# Agrego una nueva columna 'Mes' que contiene el nombre del mes correspondiente a cada fecha en español
df_final['Mes'] = df_final['release_date'].dt.month_name()
```

```python
# Reemplazo los nombres de los meses en inglés por los correspondientes en español
nombres_meses_ingles = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
nombres_meses_espanol = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
df_final['Mes'] = df_final['Mes'].replace(nombres_meses_ingles, nombres_meses_espanol)
```

```python
df_final.head(3)
```

```python
# Elimino las columnas que no voy a utilizar en las funciones para que sea más liviano el dataset
columns_to_drop = ['belongs_to_collection', 'original_language', 'overview', 'production_companies', 'production_countries', 'runtime', 'spoken_languages', 'status', 'tagline']
df_final = df_final.drop(columns=columns_to_drop)
```
