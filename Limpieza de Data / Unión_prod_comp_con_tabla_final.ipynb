```python
import pandas as pd
import ast
import numpy as np
import json
```

```python
# Ruta al archivo CSV
ruta_archivo = '/content/drive/MyDrive/NORMALIZADO prodcomp.csv'

# Cargar el DataFrame desde el archivo CSV
df = pd.read_csv(ruta_archivo)
```

```python
print(df)
```

```python
# Convertir las cadenas en diccionarios utilizando ast.literal_eval() Para la tabla normalizada de production companies
df['0'] = df['0'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else {})
```

Con esto convierto cada valor de la columna en un diccionario.


```python
##Utilizo pd.json_normalize() para desanidar los diccionarios en cada columna:
for col in df.columns:
    df_normalized = pd.json_normalize(df[col])
    df_normalized.columns = [f"{col}_{sub_col}" for sub_col in df_normalized.columns]
    df = pd.concat([df, df_normalized], axis=1)
```

```python
df_normalized.head(3)
```

```python
df.head(3)
```

```python
# Cambiar el nombre de las columnas
df = df.rename(columns={'0_name': 'production_name', '0_id': 'production_id'})
```

```python
print(df)
```

```python
# Ruta al archivo CSV abro la tabla final para agregar id y name de production
ruta_archivo = '/content/drive/MyDrive/DATAFRAME FINAL DE ARCHIVO DE MOVIES.csv'
# Cargar el DataFrame desde el archivo CSV
df_final_movies = pd.read_csv(ruta_archivo)
```

```python
df_final_movies.head(3)
```

```python
##AGREGO LA COLUMNA DE PRODUCTION_NAME A DATAFRAME FINAL
df_final_movies['production_name'] = df['production_name']
```

```python
##AGREGO LA COLUMNA DE PRODUCTION_ID A DATAFRAME FINAL
df_final_movies['production_id'] = df['production_id']
```

```python
df_final_movies.head(3)
```

```python
# Borrar las columnas que no voy a utilizar
columnas_a_borrar = ['production_companies']
df_final_movies = df_final_movies.drop(columnas_a_borrar, axis=1)
```
